```python
import numpy as np
import mne # load necessary modules
```

```python
# Determine location of sample file or data
sample_data_folder = "../files"
sample_data_file = f"{sample_data_folder}/S001/S001R01.edf"
```

```python
# Read raw data using mne
raw = mne.io.read_raw_edf(sample_data_file)
```

## Computing Power Spectral Density (PSD) and Plotting

Power Spectral Density (PSD) refers to how the power (or variance) of a signal is distributed across different frequency components.

In the context of EEG data, PSD measures the power of brain activity within specific frequency bands (such as Delta, Theta, Alpha, Beta, and Gamma bands).

These frequency bands are associated with different mental states, cognitive processes, and brain activities.

### Importance of PSD in EEG Data
*Understanding Brain states*
* Delta Waves (0.5-4 Hz): Associated with deep sleep and unconsciousness.
* Theta Waves (4-8 Hz): Often seen during drowsiness, light sleep, or relaxation
* Alpha Waves (8-13 Hz): Commonly observed during relaxed wakefulness, especially when the eyes are closed
* Beta Waves (13-30 Hz): Associated with active thinking, concentration, and motor activity
* Gamma Waves (30-100 Hz): Linked to higher cognitive functions such as memory, attention, and sensory processing

```python
raw.compute_psd(fmax=30).plot(picks="data", exclude="bads", amplitude=False)
```
